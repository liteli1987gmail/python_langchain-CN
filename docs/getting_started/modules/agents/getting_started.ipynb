# 入门

代理使用LLM来确定要执行的操作以及执行顺序。
动作可以是使用工具并观察其输出，也可以是返回给用户。

正确使用代理后，它们可以非常强大。本笔记本的目的是向您展示如何通过最简单、最高级别的API轻松使用代理。

要加载代理，您应该了解以下概念：

- 工具: 执行特定职责的函数。可以是Google搜索、数据库查找、Python REPL或其他链。工具的接口目前是预期具有字符串输入和字符串输出的函数。
- LLM：驱动代理的语言模型。
- 代理：要使用的代理。这应该是引用支持代理类的字符串。因为本笔记本专注于最简单、最高级别的API，所以仅涵盖使用标准支持的代理。如果要实现自定义代理，请参阅[自定义代理](agents/custom_agent.ipynb)的文档。

**代理**：有关支持的代理及其规格的列表，请参见[此处](agents.md)。

**工具**：有关预定义工具及其规格的列表，请参见[此处](tools.md)。

```python
from langchain.agents import load_tools
from langchain.agents import initialize_agent
from langchain.agents import AgentType
from langchain.llms import OpenAI
```

首先，让我们加载要使用的语言模型以控制代理。

```python
llm = OpenAI(temperature=0)
```

接下来，让我们加载一些要使用的工具。请注意，`llm-math`工具使用LLM，因此我们需要传递它。

```python
tools = load_tools(["serpapi", "llm-math"], llm=llm)
```

最后，让我们使用工具、语言模型以及我们要使用的代理类型来初始化代理。

```python
agent = initialize_agent(tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True)
```

现在让我们来测试一下！

```python
agent.run("Who is Leo DiCaprio's girlfriend? What is her current age raised to the 0.43 power?")
```
