# YouTube搜索工具

本笔记展示如何使用一个工具来搜索YouTube

参考自[https://github.com/venuv/langchain_yt_tools](https://github.com/venuv/langchain_yt_tools)

```python
#! pip install youtube_search
```

```python
from langchain.tools import YouTubeSearchTool
```

```python
tool = YouTubeSearchTool()
```

```python
tool.run("lex friedman")
```

您还可以指定要返回的结果数

```python
tool.run("lex friedman,5")
```
