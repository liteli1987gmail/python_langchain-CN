# Airbyte JSON（Airbyte JSON）

>[Airbyte](https://github.com/airbytehq/airbyte) 是一个数据集成平台，可以从API、数据库和文件等数据源提取数据并将其导入数据仓库和数据湖中。它拥有最大的ELT连接器目录，可连接所有数据仓库和数据库。

这里介绍如何将任何来自Airbyte的数据源导入本地JSON文件。
先决条件：安装了Docker桌面版。
步骤：
1）从GitHub上克隆Airbyte - `git clone https://github.com/airbytehq/airbyte.git`
2）进入Airbyte目录 - `cd airbyte`
3）启动Airbyte - `docker compose up`
4）在浏览器中访问http://localhost:8000。你将被要求输入用户名和密码。默认情况下，用户名为`airbyte`，密码为`password`。
5）设置想要的任何源。
6）将目标设置为本地JSON，并指定目标路径，比如`/json_data`。设置手动同步。
7）运行连接。
8）要查看创建的文件，请导航到：`file: ///tmp/airbyte_local`
9）找到你的数据并复制路径。该路径应保存在下面的文件变量中。它应该以`/tmp/airbyte_local`开头。

```python
from langchain.document_loaders import AirbyteJSONLoader
```

```python
!ls /tmp/airbyte_local/json_data/
```

```python
loader = AirbyteJSONLoader('/tmp/airbyte_local/json_data/_airbyte_raw_pokemon.jsonl')
```

```python
data = loader.load()
```
