# Azure Blob Storage容器

> [Azure Blob Storage](https://learn.microsoft.com/zh-cn/azure/storage/blobs/storage-blobs-introduction) 是微软提供的云端对象存储解决方案。Blob Storage专为存储大量非结构化数据进行了优化。非结构化数据是不遵循特定数据模型或定义的数据，例如文本或二进制数据。

`Azure Blob Storage` 专为以下目的而设计：
- 直接向浏览器提供图像或文档。
- 存储分布式访问的文件。
- 流式传输视频和音频。
- 编写日志文件。
- 存储数据以进行备份和恢复、灾难恢复和归档。
- 存储数据以供本地或Azure托管的服务分析。

本教程涵盖了如何从 `Azure Blob Storage` 容器中加载文档对象。
