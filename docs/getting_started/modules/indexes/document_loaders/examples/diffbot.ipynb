# Diffbot

>与传统的网页抓取工具不同，[Diffbot](https://docs.diffbot.com/docs)不需要任何规则来读取页面内容。
>它先从计算机视觉开始，将页面分类为20种可能的类型。然后，基于页面类型，一个机器学习模型被训练来识别页面上的关键属性。
>结果就是将网站转换为干净的结构化数据（如JSON或CSV），准备用于您的应用程序。

本文介绍如何使用[Diffbot extract API](https://www.diffbot.com/products/extract/)从一系列URL中提取HTML文档，将其转换为我们可以在下游使用的文档格式。

```python
urls = [
    "https://python.langchain.com/en/latest/index.html",
]
```

Diffbot Extract API需要API令牌。一旦拥有令牌，您就可以提取数据。

请阅读[说明](https://docs.diffbot.com/reference/authentication)以获取Diffbot API令牌。

```python
import os
from langchain.document_loaders import DiffbotLoader
loader = DiffbotLoader(urls=urls, api_token=os.environ.get("DIFFBOT_API_TOKEN"))
```

使用`.load()`方法，您可以查看已加载的文档。

```python
loader.load()
```
