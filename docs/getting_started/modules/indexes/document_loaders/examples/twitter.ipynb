# Twitter

>[Twitter](https://twitter.com/) 是一款在线社交媒体和社交网络服务。

该加载器使用 `tweepy` Python包从一批 `Twitter` 用户的推文中获取文本。
您必须使用您的 `Twitter API` 令牌初始化加载器，并传入您想要提取的 Twitter 用户名。

```python
from langchain.document_loaders import TwitterTweetLoader
```
