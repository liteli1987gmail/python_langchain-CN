# 如何序列化LLM classes

本笔记本介绍如何将LLM配置写入磁盘并从磁盘读取。如果您想保存给定LLM的配置（例如提供者、温度等），这将会很有用。

```python
from langchain.llms import OpenAI
from langchain.llms.loading import load_llm
```

## 加载
首先，让我们介绍如何从磁盘加载LLM。LLMs可以以两种格式保存在磁盘上：json或yaml。不管扩展名如何，它们的加载方式是相同的。

```python
!cat llm.json
```

```python
llm = load_llm("llm.json")
```

```python
!cat llm.yaml
```

```python
llm = load_llm("llm.yaml")
```

## 保存
如果您想要将内存中的LLM转换为序列化版本，您只需调用`.save`方法即可。同样，这也支持json和yaml。

```python
llm.save("llm.json")
```

```python
llm.save("llm.yaml")
```
