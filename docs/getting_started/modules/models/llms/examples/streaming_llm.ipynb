# 如何流式传输LLM和聊天模型响应

LangChain为LLMs提供流式传输支持。目前，我们支持`OpenAI`、`ChatOpenAI`和`ChatAnthropic`实现的流式传输，但其他LLM实现的流式传输支持在路线图上。要使用流式传输，请使用实现`on_llm_new_token`的[`CallbackHandler`](https://github.com/hwchase17/langchain/blob/master/langchain/callbacks/base.py)。在此示例中，我们使用[`StreamingStdOutCallbackHandler`]().

```python
from langchain.llms import OpenAI
from langchain.chat_models import ChatOpenAI, ChatAnthropic
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler
from langchain.schema import HumanMessage
```

```python
llm = OpenAI(streaming=True, callbacks=[StreamingStdOutCallbackHandler()], temperature=0)
resp = llm("Write me a song about sparkling water.")
```

如果使用`generate`，我们仍然可以访问结束的`LLMResult`。但目前不支持流式传输`token_usage`。

```python
llm.generate(["Tell me a joke."])
```

这是使用`ChatOpenAI`聊天模型实现的示例：

```python
chat = ChatOpenAI(streaming=True, callbacks=[StreamingStdOutCallbackHandler()], temperature=0)
resp = chat([HumanMessage(content="Write me a song about sparkling water.")])
```

这是使用`ChatAnthropic`聊天模型实现的示例，它使用他们的`claude`模型。

```python
chat = ChatAnthropic(streaming=True, callbacks=[StreamingStdOutCallbackHandler()], temperature=0)
resp = chat([HumanMessage(content="Write me a song about sparkling water.")])
```
