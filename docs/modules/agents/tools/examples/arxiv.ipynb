# ArXiv API 工具

本文展示如何使用 `arxiv` 组件。

首先，您需要安装 `arxiv` 的 Python 包。

```python
!pip install arxiv
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.agents import load_tools, initialize_agent, AgentType

llm = ChatOpenAI(temperature=0.0)
tools = load_tools(
    ["arxiv"], 
)

agent_chain = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
)
```

```python
agent_chain.run(
    "What's the paper 1605.08386 about?",
)
```

## ArXiv API 封装器

该工具封装了 API 封装器。下面，我们可以探索它提供的一些特性。

```python
from langchain.utilities import ArxivAPIWrapper
```

运行查询以获取有关一些“科学文章/文章”的信息。查询文本限制为 300 个字符。

它返回以下文章字段：
- 发布日期
- 标题
- 作者
- 摘要

下一个查询返回有关 arxiv ID 等于“1605.08386”的一篇文章的信息。

```python

arxiv = ArxivAPIWrapper()
docs = arxiv.run("1605.08386")
docs
```

现在，我们想要获取有关一位作者“Caprice Stanley”的信息。

此查询返回有关三篇文章的信息。默认情况下，查询仅返回有关三篇最佳文章的信息。

```python
docs = arxiv.run("Caprice Stanley")
docs
```

现在，我们试图查找关于不存在文章的信息。在这种情况下，响应是“未找到良好的 Arxiv 结果”。

```python
docs = arxiv.run("1605.08386WWW")
docs
```
