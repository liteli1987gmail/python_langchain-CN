# Wolfram Alpha

这个笔记本演示了如何使用Wolfram Alpha组件。

首先，您需要设置Wolfram Alpha开发人员帐户并获取您的APP ID：

1. 前往Wolfram Alpha并注册开发人员帐户[这里](https://developer.wolframalpha.com/)
2. 创建应用并获取APP ID
3. pip安装wolframalpha

然后，我们需要设置一些环境变量：
1. 将您的APP ID保存到WOLFRAM_ALPHA_APPID环境变量中
