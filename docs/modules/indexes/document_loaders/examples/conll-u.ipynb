# CoNLL-U

>[CoNLL-U](https://universaldependencies.org/format.html) 是 CoNLL-X 格式的修订版本。注释以纯文本文件形式编码（UTF-8，规范化为 NFC，仅使用 LF 字符作为换行符，包括文件末尾的 LF 字符），包含三种类型的行：
>- 在 10 个由单个制表符分隔的字段中包含单词/标记注释的单词行；请参见下面的内容。
>- 标记句子界限的空行。
>- 以井号（#）开头的注释行。

这是如何在 [CoNLL-U](https://universaldependencies.org/format.html) 格式中加载文件的示例。整个文件被视为一个文档。这个例子数据（`conllu.conllu`）基于标准的 UD/CoNLL-U 范例之一。

```python
from langchain.document_loaders import CoNLLULoader
```

```python
loader = CoNLLULoader("example_data/conllu.conllu")
```

```python
document = loader.load()
```
