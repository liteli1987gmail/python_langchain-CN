# 邮件

本笔记本展示了如何加载邮件（`.eml`）或 `Microsoft Outlook`（`.msg`） 文件。

## 使用 Unstructured（未结构化）

```python
#!pip install unstructured
```

```python
from langchain.document_loaders import UnstructuredEmailLoader
```

```python
loader = UnstructuredEmailLoader('example_data/fake-email.eml')
```

```python
data = loader.load()
```

```python
data
```

### 保留元素

在底层，Unstructured 为不同的文本块创建不同的“元素”。默认情况下，我们将它们组合在一起，但您可以通过指定 `mode="elements"` 来轻松保留分离。

```python
loader = UnstructuredEmailLoader('example_data/fake-email.eml', mode="elements")
```

```python
data = loader.load()
```

```python
data[0]
```
