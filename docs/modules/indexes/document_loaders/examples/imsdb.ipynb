# IMSDb

>[IMSDb](https://imsdb.com/)是`Internet Movie Script Database`。

本文介绍如何将`IMSDb`网页加载为文档格式，以便我们在下游使用。

```python
from langchain.document_loaders import IMSDbLoader
```

```python
loader = IMSDbLoader("https://imsdb.com/scripts/BlacKkKlansman.html")
```

```python
data = loader.load()
```
