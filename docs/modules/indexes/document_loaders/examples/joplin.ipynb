# Joplin

>[Joplin](https://joplinapp.org/) 是一个开源的笔记应用程序。可以捕捉你的想法并从任何设备安全地访问它们。

该笔记本详细介绍了如何从 `Joplin` 数据库加载文档。

`Joplin` 具有用于访问其本地数据库的 [REST API](https://joplinapp.org/api/references/rest_api/)。这个加载程序使用 API 检索数据库中的所有笔记及其元数据。这需要一个访问令牌，可以按照以下步骤从应用程序中获取：

1. 打开 `Joplin` 应用程序。必须在文档加载过程中保持该应用程序打开状态。
2. 转到设置 / 选项并选择“网络剪贴板”。
3. 确保启用了网络剪贴板服务。
4. 在“高级选项”下，复制授权令牌。

您可以直接使用访问令牌初始化加载程序，或将其存储在环境变量 JOPLIN_ACCESS_TOKEN 中。

除此之外，另一个选择是将 `Joplin` 的笔记数据库导出为Markdown文件（可选带有Front Matter元数据），然后使用MarkDown加载器，如ObsidianLoader等，来加载文件。

```python
from langchain.document_loaders import JoplinLoader
```
