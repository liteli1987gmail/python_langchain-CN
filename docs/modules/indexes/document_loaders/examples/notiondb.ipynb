# Notion DB 2/2

>[Notion](https://www.notion.so/) 是一个协作平台，支持修改后的 Markdown，并集成了看板、任务、维基和数据库。它是一个用于记笔记、知识和数据管理以及项目和任务管理的一体化工作区。

`NotionDBLoader`是一个用于从`Notion`数据库加载内容的Python类。它从数据库中检索页面，读取其内容，并返回一个文档对象的列表。

## 要求

- `Notion`数据库
- Notion集成令牌

## 设置

### 1. 创建一个Notion表数据库
在Notion中创建一个新的表数据库，您可以将任何列添加到数据库并将其视为元数据。例如，您可以添加以下列：

- 标题：将标题设置为默认属性。
- 类别：用于存储与页面相关联的类别的多选属性。
- 关键字：用于存储与页面相关联的关键字的多选属性。

将您的内容添加到数据库中每个页面的正文中。 NotionDBLoader将从这些页面提取内容和元数据。

## 2. 创建Notion集成
要创建Notion集成，请按照以下步骤操作：

1. 访问[Notion开发人员](https://www.notion.com/my-integrations) 页面，并用您的Notion帐户登录。
2. 单击“+ 新集成”按钮。
3. 为您的集成命名并选择存储数据库的工作区。
4. 选择所需的功能，此扩展名仅需要读取内容功能
5. 单击“提交”按钮以创建集成。
创建集成后，您将获得一个“集成令牌（API密钥）”。复制此令牌并妥善保管，因为您需要使用NotionDBLoader。

### 3. 将集成连接到数据库
要将集成连接到数据库，请按照以下步骤操作：

1. 在Notion中打开您的数据库。
2. 单击数据库视图右上角的三个点菜单图标。
3. 单击“+ 新集成”按钮。
4. 找到您的集成，您可能需要在搜索框中开始输入其名称。
5. 单击“连接”按钮以将集成连接到数据库。


### 4. 获取数据库ID
要获取数据库ID，请按照以下步骤操作：

1. 在Notion中打开您的数据库。
2. 单击数据库视图右上角的三个点菜单图标。
3. 从菜单中选择“复制链接”以将数据库URL复制到剪贴板中。
4. 数据库ID是在URL中找到的一长串字母数字字符。它通常看起来像这样：https://www.notion.so/username/8935f9d140a04f95a872520c4f123456?v=.... 在此示例中，数据库ID是8935f9d140a04f95a872520c4f123456。

只要正确设置了数据库并获得了集成令牌和数据库ID，就可以使用NotionDBLoader代码从Notion数据库加载内容和元数据。

## 用法
NotionDBLoader是langchain包的文档加载器的一部分。您可以按以下方式使用它：

```python
from getpass import getpass
NOTION_TOKEN = getpass()
DATABASE_ID = getpass()
```

```python
from langchain.document_loaders import NotionDBLoader
```

```python
loader = NotionDBLoader(
    integration_token=NOTION_TOKEN, 
    database_id=DATABASE_ID,
    request_timeout_sec=30 # optional, defaults to 10
)
```

```python
docs = loader.load()
```

```python
print(docs)
```
