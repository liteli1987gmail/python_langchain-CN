# Obsidian

>[Obsidian](https://obsidian.md/) 是一个功能强大、可扩展的知识库，
它基于您本地的纯文本文件夹而工作。

本手册介绍了如何从 `Obsidian` 数据库加载文档。

由于 `Obsidian` 只是以 Markdown 文件夹的形式存储在磁盘上，因此加载器只需提供对此目录的路径。

`Obsidian` 文件有时还包含元数据（请参阅 [metadata](https://help.obsidian.md/Editing+and+formatting/Metadata) ），它是文件顶部的 YAML 块。这些值将添加到文档的元数据中。（还可以向 `ObsidianLoader` 传递 `collect_metadata=False` 参数以禁用此行为。）

```python
from langchain.document_loaders import ObsidianLoader
```

```python
loader = ObsidianLoader("<path-to-obsidian>")
```

```python
docs = loader.load()
```
