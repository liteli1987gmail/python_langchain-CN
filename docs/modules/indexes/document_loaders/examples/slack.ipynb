# Slack

>[Slack](https://slack.com/)是一款即时通讯程序。

本笔记本涵盖了如何从`Slack`导出生成的Zip文件中加载文档的方法。

为了获得此“Slack”导出，按照以下步骤操作：

## 🧑 摄取您自己的数据集的说明

导出您的Slack数据。您可以通过转到Workspace管理页面并单击导入/导出选项（{your_slack_domain}.slack.com/services/export）来执行此操作。然后，选择适当的日期范围并单击“开始导出”。Slack会在导出准备就绪时向您发送电子邮件和DM。

下载将在您的下载文件夹（或根据您的操作系统配置，可以找到您的下载文件的任何位置）中产生一个`.zip`文件。

将`.zip`文件的路径复制，然后将其分配为下面的`LOCAL_ZIPFILE`。


```python
from langchain.document_loaders import SlackDirectoryLoader 
```

```python
# Optionally set your Slack URL. This will give you proper URLs in the docs sources.
SLACK_WORKSPACE_URL = "https://xxx.slack.com"
LOCAL_ZIPFILE = "" # Paste the local path to your Slack zip file here.

loader = SlackDirectoryLoader(LOCAL_ZIPFILE, SLACK_WORKSPACE_URL)
```

```python
docs = loader.load()
docs
```
