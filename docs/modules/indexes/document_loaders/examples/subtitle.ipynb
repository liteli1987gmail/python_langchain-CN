# 字幕

>[SubRip文件格式](https://en.wikipedia.org/wiki/SubRip#SubRip_file_format)被描述为“可能是所有字幕格式中最基本的”。 `SubRip（SubRip文本）`文件的扩展名为`.srt`，包含由一行空白分隔的格式化纯文本组。字幕按顺序编号，从1开始。使用的时间码格式为小时：​​分钟：​​秒，毫秒，时间单位为两个零填充的数字，分数为三个零填充的数字（00:00:00,000）。使用逗号作为小数分隔符，因为该程序是在法国编写的。

如何从字幕（`.srt`）文件加载数据

请从此处[下载示例.srt文件](https://www.opensubtitles.org/en/subtitles/5575150/star-wars-the-clone-wars-crisis-at-the-heart-en)。

```python
!pip install pysrt
```

```python
from langchain.document_loaders import SRTLoader
```

```python
loader = SRTLoader("example_data/Star_Wars_The_Clone_Wars_S06E07_Crisis_at_the_Heart.srt")
```

```python
docs = loader.load()
```
