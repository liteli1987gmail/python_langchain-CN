# 金属

>[Metal](https://github.com/getmetal/metal-python)是一个用于ML嵌入的管理服务。

这个笔记本演示如何使用[Metal](https://docs.getmetal.io/introduction)的检索器。

首先，您需要注册Metal并获得API密钥。 您可以在此处执行此操作[here](https://docs.getmetal.io/misc-create-app)。

```python
# !pip install metal_sdk
```

```python
from metal_sdk.metal import Metal
API_KEY = ""
CLIENT_ID = ""
INDEX_ID = ""

metal = Metal(API_KEY, CLIENT_ID, INDEX_ID);
```

## 摄取文档

只有在您尚未设置索引时才需要执行此操作

```python
metal.index( {"text": "foo1"})
metal.index( {"text": "foo"})
```

## 查询

现在我们的索引已设置好，我们可以设置一个检索器并开始查询它。

```python
from langchain.retrievers import MetalRetriever
```

```python
retriever = MetalRetriever(metal, params={"limit": 2})
```

```python
retriever.get_relevant_documents("foo1")
```
