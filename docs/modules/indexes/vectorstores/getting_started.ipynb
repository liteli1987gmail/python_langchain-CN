#  开始使用

此 notebook 展示了与 VectorStores 相关的基本功能。处理 vectorstores 的关键部分是创建向量以放入向量库中，通常通过嵌入来创建。因此，在深入本文之前，建议您首先熟悉 [embedding notebook](../../models/text_embedding.htpl)。

此文涵盖了所有向量库相关的通用高级功能。

```python
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Chroma
```

```python
with open('../../state_of_the_union.txt') as f:
    state_of_the_union = f.read()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
texts = text_splitter.split_text(state_of_the_union)

embeddings = OpenAIEmbeddings()
```

```python
docsearch = Chroma.from_texts(texts, embeddings)

query = "What did the president say about Ketanji Brown Jackson"
docs = docsearch.similarity_search(query)
```

```python
print(docs[0].page_content)
```

## 添加文本
您可以使用 `add_texts` 方法轻松地将文本添加到向量库中。它将返回文档 ID 的列表（如果您需要在下游使用它们）。

```python
docsearch.add_texts(["Ankush went to Princeton"])
```

```python
query = "Where did Ankush go to college?"
docs = docsearch.similarity_search(query)
```

```python
docs[0]
```

## 来自文档
我们还可以直接从文档初始化向量库。当原始文档有关联的元数据时，使用文本分割器方法直接获取文档很方便。

```python
documents = text_splitter.create_documents([state_of_the_union], metadatas=[{"source": "State of the Union"}])
```

```python
docsearch = Chroma.from_documents(documents, embeddings)

query = "What did the president say about Ketanji Brown Jackson"
docs = docsearch.similarity_search(query)
```

```python
print(docs[0].page_content)
```
