# Momento聊天消息历史记录

本笔记本讲述了如何使用[Momento Cache]（https://gomomento.com）使用`MomentoChatMessageHistory`类存储聊天消息历史记录。请参阅Momento[docs]（https://docs.momentohq.com/getting-started）以获取有关如何使用Momento的更多详细信息。

请注意，默认情况下，如果不存在具有给定名称的缓存，则将创建缓存。

您需要获取Momento身份验证令牌才能使用此类。这可以通过将其传递到momento.CacheClient中直接实例化，作为`auth_token`的命名参数传递给`MomentoChatMessageHistory.from_client_params`，或者可以将其设置为环境变量`MOMENTO_AUTH_TOKEN`来完成。# 

```python
from datetime import timedelta

from langchain.memory import MomentoChatMessageHistory

session_id = "foo"
cache_name = "langchain"
ttl = timedelta(days=1)
history = MomentoChatMessageHistory.from_client_params(
    session_id, 
    cache_name,
    ttl,
)

history.add_user_message("hi!")

history.add_ai_message("whats up?")
```
